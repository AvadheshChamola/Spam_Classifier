```python
dataset_path="/content/drive/MyDrive/Colab Notebooks/NLP/smsspamcollection/SMSSpamCollection"
```

```python
import pandas as pd
dataset=pd.read_csv(dataset_path,sep='\t',names=["label","message"])
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger') 
```

```python
dataset
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

```python
scorpus=[]
for i in range(len(dataset)):
  review=re.sub('[^a-zA-Z]',' ',dataset['message'][i])
  review=review.lower().split()
  review=[stemmer.stem(word) for word in review if word not in set(stopwords.words('english'))]
  review=' '.join(review)
  scorpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=3000)
```

```python
Xs=cv.fit_transform(scorpus).toarray()
```

```python
y=pd.get_dummies(dataset['label'])
```

```python
y
```

```python
y=y.iloc[:,1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xs,y,test_size=0.20,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion=confusion_matrix(y_test,y_pred)
```
